```python
#Convert issue_date to datetime
data_v4['issue_d'] = pd.to_datetime(data_v4['issue_d'])
```

```python
data_v5 = data_v4.drop(['loan_status_Charged Off'], axis=1)
data_v5
```

```python
# Double check no 'na' in dataset
data_v5.isna().sum()
```

```python
## Machine Learning Model #1 ##
```

```python
#Setting y/prediction variable column
y = data_v5['loan_status_Fully Paid']
#y = y.reset_index()
#y = y.drop(columns='index')
y.ravel()
y
```

```python
#Setting x/dependent variables 
X = data_v5[[
'annual_inc','fico_range_low', 'dti', 'loan_amnt',
'open_acc',
'pub_rec',
'pub_rec_bankruptcies',
'tax_liens', 'home_ownership_ANY',
'home_ownership_MORTGAGE',
'home_ownership_NONE',
'home_ownership_OTHER',
'home_ownership_OWN',
'home_ownership_RENT']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score



```

```python
# split Data into training and testing sets **USING STRATIFY method**
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
#Instantiate Logistic Regression Model   - KEEP random_state as 1 -
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
   intercept_scaling=1, l1_ratio=None, max_iter=100,
   multi_class='warn', n_jobs=None, penalty='12',
   random_state=1, solver='lbfgs', tol=0.0001, verbose=0,
   warm_start=False)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
print(X_train.shape, y_train.shape)
```

```python
y_train
```

```python
#Training the Logistic Regression model
classifier.fit(X_train, y_train)
```

```python
# Validate the Logistic Regression Model
predictions = classifier.predict(X_test)
```

```python
#Accuracy Score
accuracy_score(y_test, predictions)
```

```python
y_test, predictions
```

```python
output_df = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
output_df
```

```python
from sklearn.metrics import confusion_matrix, classification_report
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
report = classification_report(y_test, predictions)
print(report)
```
